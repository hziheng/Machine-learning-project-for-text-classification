```python
import re
def get_data(path, label):
    datas = []
    labels = []
    with open(path, 'r', encoding='utf-8') as ff:
        for i in ff:
            if i.strip():
                i = i.rstrip()
                labels.append(label[i[-1]])
                data = re.sub(' +', ',', i[:-1])
                datas.append(data.strip())
    return datas, labels
            
```

```python
test_data, test_label = get_data('./data/dl_data/test.txt', label)
```

```python
train_data, train_label = get_data('./data/dl_data/train.txt', label)
```

```python
dev_data, dev_label = get_data('./data/dl_data/dev.txt', label)
```

```python
def create_csv(datax, datay, save_path):
    df = pd.DataFrame()
    df['content'] = datax
    df['label'] = datay
    df.to_csv(save_path, index=False, encoding='utf-8')
```

```python
create_csv(test_data, test_label, './data/dl_data/test.csv')
create_csv(train_data, train_label, './data/dl_data/train.csv')
create_csv(dev_data, dev_label, './data/dl_data/dev.csv')
```
